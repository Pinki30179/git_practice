```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def rollDie(num):
    rolls=[0]*6
    for i in range(0,num):
        roll=int(np.random.randint(1,6))
        rolls[roll-1]+=1
    return rolls
```

```python
def main():
      ran_rools=10000
      result=rollDie(ran_rools)
      print(result)
      samp=np.array([1,2,3,4,5,6])
      plt.ylabel(result)
      sns.histplot(samp,stat='probability', bins=12);
```

```python
if __name__=='__main__':
    main()
```
